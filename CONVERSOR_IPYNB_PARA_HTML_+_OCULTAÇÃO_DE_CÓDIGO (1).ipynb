# **üõ†Ô∏è Manual de Uso do Conversor**

Esta ferramenta foi desenvolvida para transformar notebooks de an√°lise de dados (`.ipynb`) em relat√≥rios HTML profissionais, prontos para entrega acad√™mica ou publica√ß√£o.

### **O que ela faz:**
1. Converte seu notebook Jupyter/Colab para HTML.
2. Injeta um script que **oculta automaticamente as c√©lulas de c√≥digo**, mantendo o foco na narrativa e nos gr√°ficos.
3. Adiciona bot√µes interativos ("Ver C√≥digo") caso o avaliador queira inspecionar a implementa√ß√£o t√©cnica.

### **üìù Passo a Passo**

1. **Baixe seu Trabalho:**
   No notebook da sua an√°lise (ex: *Trabalho_Final.ipynb*), v√° em `Arquivo` > `Fazer Download` > `Fazer download do .ipynb`.

2. **Execute o Conversor:**
   Clique no bot√£o de **Play (‚ñ∂Ô∏è)** na c√©lula de c√≥digo abaixo.

3. **Fa√ßa o Upload:**
   Um bot√£o aparecer√° na tela. Clique nele e selecione o arquivo `.ipynb` que voc√™ baixou no passo 1.

4. **Aguarde o Processamento:**
   O sistema far√° a convers√£o e limpeza. Em poucos segundos, o navegador iniciar√° o download autom√°tico do arquivo `Relatorio_Final_....html`.


>**Nota:** O arquivo gerado √© independente e pode ser aberto em qualquer navegador sem necessidade de internet ou Python instalado.


---



```python
# @title
import os
from google.colab import files
from IPython.display import display, HTML

display(HTML("""
<div style="font-family: 'Roboto', 'Helvetica Neue', Arial, sans-serif; max-width: 650px; margin: 20px auto; border-radius: 8px; box-shadow: 0 10px 25px rgba(0,0,0,0.1); overflow: hidden; border: 1px solid #e1e4e8;">
    <div style="background-color: #24292e; padding: 25px; text-align: center; border-bottom: 4px solid #f9826c;">
        <h1 style="color: white; margin: 0; font-weight: 700; letter-spacing: 1.5px; font-size: 22px;">CONVERSOR</h1>
        <p style="color: #959da5; margin: 8px 0 0 0; font-size: 13px;">Notebook (.ipynb) ‚ûî HTML Interativo</p>
    </div>
    <div style="padding: 30px; background-color: #ffffff;">
        <div style="border: 2px dashed #d1d5da; border-radius: 6px; padding: 25px; text-align: center; background-color: #f6f8fa; transition: all 0.3s;">
            <p style="color: #24292e; font-size: 15px; margin: 0; font-weight: 500;">
                Clique no bot√£o abaixo para selecionar seu arquivo.
            </p>
            <p style="color: #586069; font-size: 12px; margin: 8px 0 0 0;">
                Oculta√ß√£o autom√°tica de c√≥digo e preserva√ß√£o de textos.
            </p>
        </div>
    </div>
</div>
"""))

uploaded = files.upload()

if uploaded:
    notebook_filename = list(uploaded.keys())[0]

    if not notebook_filename.lower().endswith('.ipynb'):
        print("Erro: O arquivo deve ter extens√£o .ipynb")
    else:
        safe_filename = f'"{notebook_filename}"'
        os.system(f'jupyter nbconvert --to html {safe_filename}')

        html_filename = notebook_filename.replace('.ipynb', '.html')

        try:
            with open(html_filename, "r", encoding='utf-8') as f:
                html_content = f.read()

            script_injection = """
            <style>
                .code-toggle-bar {
                    background-color: #f6f8fa;
                    border: 1px solid #e1e4e8;
                    border-bottom: none;
                    border-radius: 6px 6px 0 0;
                    padding: 8px 12px;
                    display: flex;
                    align-items: center;
                    justify-content: flex-end;
                    font-family: -apple-system, BlinkMacSystemFont, "Segoe UI", Helvetica, Arial, sans-serif;
                    margin-top: 15px;
                }

                .toggle-btn {
                    background-color: #ffffff;
                    border: 1px solid #d1d5da;
                    border-radius: 6px;
                    color: #24292e;
                    padding: 4px 12px;
                    font-size: 12px;
                    font-weight: 600;
                    cursor: pointer;
                    display: flex;
                    align-items: center;
                    gap: 6px;
                    transition: all 0.2s cubic-bezier(0.3, 0, 0.5, 1);
                    outline: none;
                    box-shadow: 0 1px 0 rgba(27,31,35,0.04);
                }

                .toggle-btn:hover {
                    background-color: #f3f4f6;
                    border-color: #c0c4c8;
                    text-decoration: none;
                }

                .toggle-btn.active {
                    background-color: #eff3f6;
                    border-color: #0366d6;
                    color: #0366d6;
                }

                .code_cell .input,
                .jp-CodeCell .jp-Cell-inputWrapper,
                .jp-CodeCell .input_area {
                    display: none !important;
                    opacity: 0;
                    transition: opacity 0.3s ease;
                }

                .show-code .input,
                .show-code .jp-Cell-inputWrapper,
                .show-code .input_area {
                    display: block !important;
                    opacity: 1;
                    border: 1px solid #e1e4e8;
                    border-radius: 0 0 6px 6px;
                }

                .text_cell,
                .jp-MarkdownCell,
                .jp-RenderedHTMLCommon,
                .text_cell_render {
                    display: block !important;
                    visibility: visible !important;
                    opacity: 1 !important;
                }
            </style>

            <script>
            document.addEventListener("DOMContentLoaded", function() {
                var selectors = '.code_cell, .jp-CodeCell';
                var cells = document.querySelectorAll(selectors);

                cells.forEach(function(cell) {
                    if (cell.classList.contains('text_cell') || cell.classList.contains('jp-MarkdownCell')) {
                        return;
                    }

                    var inputBlock = cell.querySelector('.input, .input_area, .jp-Cell-inputWrapper');
                    if (!inputBlock) return;

                    var bar = document.createElement('div');
                    bar.className = 'code-toggle-bar';

                    var btn = document.createElement('button');
                    btn.className = 'toggle-btn';
                    btn.innerHTML = '<span>üëÅÔ∏è</span> Ver C√≥digo';

                    btn.onclick = function() {
                        cell.classList.toggle('show-code');

                        if (cell.classList.contains('show-code')) {
                            btn.innerHTML = '<span>‚úï</span> Ocultar';
                            btn.classList.add('active');
                            bar.style.borderBottom = '1px solid #e1e4e8';
                        } else {
                            btn.innerHTML = '<span>üëÅÔ∏è</span> Ver C√≥digo';
                            btn.classList.remove('active');
                            bar.style.borderBottom = 'none';
                        }
                    };

                    bar.appendChild(btn);
                    cell.insertBefore(bar, inputBlock);
                });
            });
            </script>
            """

            if "</body>" in html_content:
                final_html = html_content.replace("</body>", script_injection + "</body>")
            else:
                final_html = html_content + script_injection

            output_name = "Relatorio_Final_" + html_filename
            with open(output_name, "w", encoding='utf-8') as f:
                f.write(final_html)

            display(HTML(f"""
            <div style="max-width: 650px; margin: 20px auto; background-color: #f0fff4; color: #22863a; padding: 15px; border-radius: 6px; border: 1px solid #c3e6cb; text-align: center; font-family: sans-serif; font-size: 14px;">
                <strong style="display: block; margin-bottom: 4px;">‚úî Processo Conclu√≠do</strong>
                O arquivo <em>{output_name}</em> foi gerado e ser√° baixado agora.
            </div>
            """))

            files.download(output_name)

        except Exception as e:
            print(f"Erro: {str(e)}")
```
